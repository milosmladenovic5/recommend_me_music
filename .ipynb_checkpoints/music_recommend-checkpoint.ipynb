```python
users_dictionary = {}
items_dictionary = {}
ci = 0 
cu = 0

for i in range(0, music_rat_df.shape[0]-1):
    if (items_dictionary.get(music_rat_df.iloc[i].artist_id) == None):
        items_dictionary.update({ music_rat_df.iloc[i].artist_id : ci})
        ci+=1
        
    if (users_dictionary.get(music_rat_df.iloc[i].reviewer_id) == None):
        users_dictionary.update({music_rat_df.iloc[i].reviewer_id : cu})
        cu+=1
```

```python
item_val_array = np.fromiter(iter(items_dictionary.values()), dtype=int)
item_val_array = np.sort(item_val_array)

user_val_array = np.fromiter(iter(users_dictionary.values()), dtype = int)
user_val_array = np.sort(user_val_array)
```

```python
train_data_matrix = np.zeros((n_unique_users, n_unique_items))
for line in train_data.itertuples():
    train_data_matrix[users_dictionary.get(line[4]), items_dictionary.get(line[1])] = line[2]
    
test_data_matrix = np.zeros((n_unique_users, n_unique_items))
for line in test_data.itertuples():
    test_data_matrix[users_dictionary.get(line[4]), items_dictionary.get(line[1])] = line[2]
```

# Kolaborativno filtriranje

- <h4> Prvo računamo sličnost među korisnicima </h4>
<p> U tu svrhu, koristimo sledeću formulu: </p>
<img class="aligncenter size-thumbnail img-responsive" src="https://latex.codecogs.com/gif.latex?s_u^{cos}(u_k,u_a)=\frac{u_k&space;\cdot&space;u_a&space;}{&space;\left&space;\|&space;u_k&space;\right&space;\|&space;\left&space;\|&space;u_a&space;\right&space;\|&space;}&space;=\frac{\sum&space;x_{k,m}x_{a,m}}{\sqrt{\sum&space;x_{k,m}^2\sum&space;x_{a,m}^2}}"/>

```python
user_similarity = pairwise_distances(train_data_matrix, metric = 'cosine')
```

 - <h4> Zatim računamo sličnost između izvođača čiji su albumi prodavani </h4>
 <p> U tu svrhu, koristimo sledeću formulu:</p>
 
 <img class="aligncenter size-thumbnail img-responsive" src="https://latex.codecogs.com/gif.latex?s_a^{cos}(a_m,a_b)=\frac{a_m&space;\cdot&space;a_b&space;}{&space;\left&space;\|&space;a_m&space;\right&space;\|&space;\left&space;\|&space;a_b&space;\right&space;\|&space;}&space;=\frac{\sum&space;x_{i,m}x_{i,b}}{\sqrt{\sum&space;x_{i,m}^2\sum&space;x_{i,b}^2}}
"/>

```python
item_similarity = pairwise_distances(train_data_matrix.T, metric = 'cosine')
```

# Kolaborativno filtriranje sa uklonjenom pristrasnošću (bias-substracted)

- U sledećem koraku pravimo predikciju ocena. Već smo kreirali matrice sličnosti "user_similarity" i "item_similarity", pa stoga možemo da vršimo predikciju primenjujući sledeću formulu za kolaborativno filtriranje bazirano na korisnicima:

<img class="aligncenter size-thumbnail img-responsive" src="https://latex.codecogs.com/gif.latex?\hat{x}_{k,m}&space;=&space;\bar{x}_{k}&space;&plus;&space;\frac{\sum\limits_{u_a}&space;sim_u(u_k,&space;u_a)&space;(x_{a,m}&space;-&space;\bar{x_{u_a}})}{\sum\limits_{u_a}|sim_u(u_k,&space;u_a)|}"/>

Možemo da gledamo na sličnost između korisnika k i a kao težine kojima se množe ocene sličnih korisnika (ispravljene za prosečni rejting tog korisnika). Potrebno je da normalizujemo te ocene tako da ostanu između 1 i 5 i, kao poslednji korak, sumiramo prosečne ocene. Obzirom na to da neki korisnici teže tome da daju jako dobre ili jako loše ocene, uprosečavanjem njihovih ocena uklanjamo tu njihovu "pristrasnost".



- Kada pravimo predikciju za kolaborativno filtriranje bazirano na objektu, nije potrebno da ispravljamo za prosečni rejting korisnika, zato što se sami korisnici koriste za predikciju.


<img class="aligncenter size-thumbnail img-responsive" src="https://latex.codecogs.com/gif.latex?\hat{x}_{k,m}&space;=&space;\frac{\sum\limits_{i_b}&space;sim_i(i_m,&space;i_b)&space;(x_{k,b})&space;}{\sum\limits_{i_b}|sim_i(i_m,&space;i_b)|}"/>

```python
def predict_non_biased(ratings, similarity, kind = 'user'):
    if kind == 'user':
        mean_user_rating = np.nanmean(ratings, axis = 1)
        #You use np.newaxis so that mean_user_rating has same format as ratings
        ratings_diff = (ratings - mean_user_rating[:, np.newaxis])
        pred = mean_user_rating[:, np.newaxis] + similarity.dot(ratings_diff) / np.array([np.abs(similarity).sum(axis = 1)]).T
    elif kind == 'item':
        pred = ratings.dot(similarity) / np.array([np.abs(similarity).sum(axis = 1)])
    return pred
```

```python
user_prediction_non_biased = predict_non_biased(train_data_matrix, user_similarity, kind = 'user')
item_prediction_non_biased = predict_non_biased(train_data_matrix, item_similarity, kind = 'item')
```

# Kolaborativno filtiranje po k- najsličnijih korisnika, sa uklonjenom pristrasnošću

- ideja je da se sada više za predikciju ne koriste svi korisnici iz skupa podataka, bez obzira na njihovu sličnost sa onim za kog se vrši predviđanje, nego k najsličnijih korisnika i da se kasnije ispita greška

```python
def predict_topk_non_biased(ratings, similarity, kind='user', k=40):
    pred = np.zeros(ratings.shape)
    if kind == 'user':
        user_bias = ratings.mean(axis=1)
        ratings = (ratings - user_bias[:, np.newaxis]).copy()
        for i in range(ratings.shape[0]):
            top_k_users = [np.argsort(similarity[:,i])[:-k-1:-1]]
            for j in range(ratings.shape[1]):
                pred[i, j] = similarity[i, :][top_k_users].dot(ratings[:, j][top_k_users]) 
                pred[i, j] /= np.sum(np.abs(similarity[i, :][top_k_users]))
        pred += user_bias[:, np.newaxis]
    if kind == 'item':
        item_bias = ratings.mean(axis=0)
        ratings = (ratings - item_bias[np.newaxis, :]).copy()
        for j in range(ratings.shape[1]):
            top_k_items = [np.argsort(similarity[:,j])[:-k-1:-1]]
            for i in range(ratings.shape[0]):
                pred[i, j] = similarity[j, :][top_k_items].dot(ratings[i, :][top_k_items].T) 
                pred[i, j] /= np.sum(np.abs(similarity[j, :][top_k_items])) 
        pred += item_bias[np.newaxis, :]
        
    return pred
```

```python
topk_user_pred_non_biased = predict_topk_non_biased(train_data_matrix, user_similarity, kind = 'user', k = 15)
topk_item_pred_non_biased = predict_topk_non_biased(train_data_matrix, item_similarity, kind = 'item', k = 15)
```

- <h3> Evaluacija modela koristeći MSE (mean squared error) - srednju kvadratnu grešku : </h3>
<img src="https://latex.codecogs.com/gif.latex?MSE&space;=\frac{1}{N}&space;\sum&space;(x_i&space;-\hat{x_i})^2" title="MSE =\sqrt{\frac{1}{N} \sum (x_i -\hat{x_i})^2}" />

```python
def mse(pred, actual):
    # Ignore nonzero terms.
    pred = pred[actual.nonzero()].flatten()
    actual = actual[actual.nonzero()].flatten()
    return mean_squared_error(pred, actual)
```

```python
print('User-based CF MSE: ' + str(mse(user_prediction_non_biased, test_data_matrix)))
print('Item-based CF MSE: ' + str(mse(item_prediction_non_biased, test_data_matrix)))
```

```python
print('Top K User-based CF MSE: ' + str(mse(topk_user_pred_non_biased, test_data_matrix)))
print('Top K Item-based CF MSE: ' + str(mse(topk_item_pred_non_biased, test_data_matrix)))
```

# Kolaborativno filtriranje sa pristrasnošću (biased)

- Upotrebićemo formulu sličnu gornjoj, samo ovog puta nećemo vršiti korekciju za prosečnu ocenu

<img class="aligncenter size-thumbnail img-responsive" src="https://latex.codecogs.com/gif.latex?\hat{x}_{k,m}&space;=&space;&space;&space;\frac{\sum\limits_{u_a}&space;sim_u(u_k,&space;u_a)&space;(x_{a,m})}{\sum\limits_{u_a}|sim_u(u_k,&space;u_a)|}"/>

- Kolaborativno filtriranje zasnovano na objektima (item-based) zapravo ima istu formulu i u ovom slučaju, odnosno nema potrebe za uklanjanjem ili dodavanjem pristrasnosti, jer ona ne postoji


```python
def predict_biased(ratings, similarity, kind='user'):
    if kind == 'user':
        return similarity.dot(ratings) / np.array([np.abs(similarity).sum(axis=1)]).T
    elif kind == 'item':
        return ratings.dot(similarity) / np.array([np.abs(similarity).sum(axis=1)])
```

```python
user_prediction_biased = predict_biased(train_data_matrix, user_similarity, kind = 'user')
```

# Kolaborativno filtiranje po k - najsličnijih korisnika, sa pristrasnošću (biased)

```python
def predict_topk_biased(ratings, similarity, kind='user', k=40):
    pred = np.zeros(ratings.shape)
    if kind == 'user':
        for i in range(ratings.shape[0]):
            top_k_users = [np.argsort(similarity[:,i])[:-k-1:-1]]
            for j in range(ratings.shape[1]):
                pred[i, j] = similarity[i, :][top_k_users].dot(ratings[:, j][top_k_users]) 
                pred[i, j] /= np.sum(np.abs(similarity[i, :][top_k_users]))
    if kind == 'item':
        for j in range(ratings.shape[1]):
            top_k_items = [np.argsort(similarity[:,j])[:-k-1:-1]]
            for i in range(ratings.shape[0]):
                pred[i, j] = similarity[j, :][top_k_items].dot(ratings[i, :][top_k_items].T) 
                pred[i, j] /= np.sum(np.abs(similarity[j, :][top_k_items]))        
    
    return pred
```

```python
topk_user_pred_biased = predict_topk_non_biased(train_data_matrix, user_similarity, kind = 'user', k = 15)
```

- <h2> Proveravamo performanse modela koristeći predviđene ocene da dobijemo informacije o albumima i izvođačima koji bi se određenom korisniku dopali </h2>

```python
def get_top_artist_ids(artists_dictionary, values):
    ids = []
    for i in range(0, len(values-1)):
        ids.append(list(artists_dictionary.keys())[list(artists_dictionary.values()).index(values[i])])
    
    return ids
```

```python
def get_artists_description(artist_ids, music_df):
    descriptions = []
    
    for i in range(0, len(artist_ids) - 1):
        res_df = music_df.query("artist_id == '" + artist_ids[i] + "'")
        descriptions.append(res_df['summary'].tolist())
    return descriptions
```

# Kolaborativno filtriranje bazirano na modelu
  - <b>SVD</b> (Singular Value Decomposition)
  
Dobro poznata tehnika za faktorizaciju matrica naziva se **Singular value decomposition (SVD)**. Kolaborativno filtriranje može biti formulisano kao aproksimacija matrice X korišćenjem SVD. 

Glavna jednačina izgleda ovako:
<img src="https://latex.codecogs.com/gif.latex?X=USV^T" title="X=USV^T" />


Ako matrica <b>`X`</b> ima dimenzije <b>`m x n`</b> :
* *`U`* je  *`(m x k)`* ortogonalna matrica
* *`S`* je  *`(k x k)`*  dijagonalna matrica sa ne-negativnim realnim brojevima na dijagonali
* *V^T* je *`(k x n)`* ortogonalna matrica


Matrica *`X`* može biti faktorizovana na *`U`*, *`S`* and *`V`*. *`U`* matrica predstavlja vektore odlika koji odgovaraju korisnicima u skrivenom prostoru odlika, dok  *`V`* matrica predstavlja vektore odlika  koji odgovaraju odlikama objekata u skrivenom prostoru odlika.
<img class="aligncenter size-thumbnail img-responsive" style="max-width:100%; width: 50%; max-width: none" src="http://s33.postimg.org/kwgsb5g1b/BLOG_CCA_5.png"/>

Sada možemo da predvidimo ocene za neocenjene filmove množenjem matrica *`U`*, *`S`* i *`V^T`*.

<img class="aligncenter size-thumbnail img-responsive" style="max-width:100%; width: 50%; max-width: none" src="http://s33.postimg.org/ch9lcm6pb/BLOG_CCA_4.png"/>

```python
#adjusting parameter K for SVD
error_rate = []

# Will take some time
for i in range(1,50):  
    u, s, vt = svds(train_data_matrix, k = i)
    s_diag_matrix=np.diag(s)
    X_pred = np.dot(np.dot(u, s_diag_matrix), vt)
    error_rate.append(mse(X_pred, test_data_matrix))
```

```python
for i in range(50,70):  
    u, s, vt = svds(train_data_matrix, k = i)
    s_diag_matrix=np.diag(s)
    X_pred = np.dot(np.dot(u, s_diag_matrix), vt)
    error_rate.append(mse(X_pred, test_data_matrix))
```

```python
plt.figure(figsize=(15,10))
plt.xlabel("K")
plt.ylabel("Error rate")
plt.plot(error_rate, "r")
plt.show()
```

```python
np.argmin(np.array(error_rate))
```
